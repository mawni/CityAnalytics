```python
# import 
import geopandas as gpd
import numpy as np
import pandas as pd
import fiona
import fsspec
import json
#from shapely.geometry import Point
#from shapely.geometry.polygon import Polygon
```

```python
# return a df that have two column: 1 "sa2_code" 2 "life_sat"
# assume the file "life satisfy.csv" in current directory
def life_sat_by_region():
    data = pd.read_csv("life satisfy.csv")
    sa2_code = data[" sa2_code16"]
    data = data.drop(" sa2_code16", axis = 1)
    
    # calculate the happiness score
    for i in range(0,80,10):
        data[" _life_satisfaction_pc_"+str(i)+"_synth"] = data[" _life_satisfaction_pc_"+str(i)+"_synth"] * i
    data['life_satisfaction_pc_80_to_100_synth'] = data['life_satisfaction_pc_80_to_100_synth']*90 # calculate by average
    total = data.sum(axis = 1)
    temp = pd.concat([sa2_code, total],axis = 1)
    temp.columns = ["sa2_code", "life_sat"]
    
    return temp
```

```python
# regional qualifiaction score
# calculation: advanced diploma and diploma = 5.5 
# bachelor = 7 
# certification 1 and 2 = 1.5
# certification 3 and 4 = 3.5
# not further defined = 4
# graduate certification = 8
# post degree (exc gd) = 9.5

# assume the qualification.csv is in current directory
def get_regional_education_score():
    data = pd.read_csv("qualification.csv")
    normalized_score =  (data["p_bachdeg_total"]*7 + data[" p_cert_iii_iv_total"]*3.5 + data[" p_cert_i_ii_total"]*1.5 + data[" p_graddip_and_gradcert_total"]*8 
                        + data[" p_advdip_and_dip_total"]*5.5 + (data[" p_pgrad_deg_total"] - data[" p_graddip_and_gradcert_total"]) * 9.5)/data[" p_tot_total"]
    temp = pd.concat([data[" sa2_main16"], normalized_score], axis = 1)
    temp.columns = ["sa2_code", "education_score"]
    return temp
```

```python
# main program

def main():
    # read population data
    pop = pd.read_csv("population by sa2 2016.csv")
    pop.columns = ["sa2_code", "size"]
    pop["sa2_code"] = pd.to_numeric(pop["sa2_code"])
    sorted_pop = pop.sort_values(by = ["sa2_code"])
    
    # load salary
    income = pd.read_csv("median salary.csv")[[" sa2_main16", "med_psnl_inc_wk_tot"]]
    income.columns = ["sa2_code", "weekly_income"]
    income["sa2_code"] = pd.to_numeric(income["sa2_code"])
    sorted_income = income.sort_values(by = ["sa2_code"])
    
    # load life satisfication data
    life_sat = life_sat_by_region()
    life_sat["sa2_code"] = pd.to_numeric(life_sat["sa2_code"])
    sorted_life_sat = life_sat.sort_values(by = ["sa2_code"])
    
    # load regional education level score
    education_score = get_regional_education_score()
    education_score["sa2_code"] = pd.to_numeric(education_score["sa2_code"])
    sorted_education_score = education_score.sort_values(by = ["sa2_code"])
    
    final = pd.concat([sorted_pop, sorted_income["weekly_income"], sorted_life_sat["life_sat"], sorted_education_score["education_score"]], axis = 1)
    
    return final
```

```python
a = main()
```
